```python
import random

# 1. O computador escolhe um número secreto entre 1 e 10
numero_secreto = random.randint(1, 10)

print("=========================================")
print("🎲 BEM-VINDO AO JOGO DA ADIVINHAÇÃO! 🎲")
print("=========================================")
print("Eu pensei em um número de 1 a 10. Consegue adivinhar?")

# 2. O usuário digita o palpite (convertemos o texto digitado para um número inteiro)
palpite = int(input("Digite o seu palpite: "))

# 3. O código revela o número pensado e valida o resultado
print(f"\n🤖 O número que eu pensei foi: {numero_secreto}")

if palpite == numero_secreto:
    print("🎉 Parabéns! Você tem poderes mediúnicos, você acertou!")
else:
    print("❌ Quase! (Ou não...) Tente a sorte novamente!")
```
