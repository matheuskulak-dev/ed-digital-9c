### Quiz sobre Kurt Cobain

Este quiz testará seus conhecimentos sobre Kurt Cobain. Você precisará responder a 10 perguntas, e o jogo termina se você errar alguma. Se acertar todas, você vence!

```python
import random

# Lista de perguntas sobre Kurt Cobain
questions = [
    {
        "question": "Qual é o nome da banda mais famosa de Kurt Cobain?",
        "options": {"a": "Pearl Jam", "b": "Nirvana"},
        "correct_answer": "b"
    },
    {
        "question": "Em qual cidade Kurt Cobain nasceu?",
        "options": {"a": "Aberdeen, Washington", "b": "Seattle, Washington"},
        "correct_answer": "a"
    },
    {
        "question": "Qual o nome do primeiro álbum de estúdio do Nirvana?",
        "options": {"a": "Nevermind", "b": "Bleach"},
        "correct_answer": "b"
    },
    {
        "question": "Qual o nome da esposa de Kurt Cobain?",
        "options": {"a": "Courtney Love", "b": "Frances Bean Cobain"},
        "correct_answer": "a"
    },
    {
        "question": "Qual das seguintes músicas NÃO é do Nirvana?",
        "options": {"a": "Lithium", "b": "Black Hole Sun"},
        "correct_answer": "b"
    },
    {
        "question": "Em que ano Kurt Cobain faleceu?",
        "options": {"a": "1994", "b": "1997"},
        "correct_answer": "a"
    },
    {
        "question": "Qual instrumento Kurt Cobain tocava principalmente?",
        "options": {"a": "Baixo", "b": "Guitarra"},
        "correct_answer": "b"
    },
    {
        "question": "Qual o nome da filha de Kurt Cobain?",
        "options": {"a": "Frances Bean Cobain", "b": "Lila Cobain"},
        "correct_answer": "a"
    },
    {
        "question": "Onde o Nirvana fez seu famoso show 'MTV Unplugged'?",
        "options": {"a": "Nova York", "b": "Los Angeles"},
        "correct_answer": "a"
    },
    {
        "question": "Qual o nome do álbum que contém a música 'Smells Like Teen Spirit'?",
        "options": {"a": "In Utero", "b": "Nevermind"},
        "correct_answer": "b"
    },
    {
        "question": "Antes de formar o Nirvana, Kurt Cobain estava em uma banda chamada...?",
        "options": {"a": "Fecal Matter", "b": "Melvana"},
        "correct_answer": "a"
    },
    {
        "question": "Qual desses é um apelido associado a Kurt Cobain?",
        "options": {"a": "The Lizard King", "b": "The Voice of a Generation"},
        "correct_answer": "b"
    }
]

def run_quiz():
    print("Bem-vindo ao Quiz de Kurt Cobain!")
    print("Você terá 10 perguntas e precisa acertar todas para ganhar.")
    print("Responda com 'a' ou 'b'. Se errar, o jogo acaba!\n")

    # Sortear 10 perguntas aleatoriamente
    selected_questions = random.sample(questions, 10)
    score = 0

    for i, q_data in enumerate(selected_questions):
        print(f"Pergunta {i + 1}: {q_data['question']}")
        print(f"A) {q_data['options']['a']}")
        print(f"B) {q_data['options']['b']}")

        user_answer = input("Sua resposta (a/b): ").lower().strip()

        while user_answer not in ['a', 'b']:
            print("Resposta inválida. Por favor, digite 'a' ou 'b'.")
            user_answer = input("Sua resposta (a/b): ").lower().strip()

        if user_answer == q_data['correct_answer']:
            print("Correto!\n")
            score += 1
        else:
            print(f"Errado! A resposta correta era {q_data['correct_answer'].upper()}) {q_data['options'][q_data['correct_answer']]}.")
            print("Game Over. Você perdeu.\n")
            return # Termina o jogo

    if score == 10:
        print("Parabéns! Você acertou todas as 10 perguntas e venceu o jogo!")
    else:
        # Este caso não deve ser atingido se o 'return' acima funcionar corretamente.
        # Mas é um fallback para garantir que sempre há uma mensagem final.
        print("Obrigado por jogar!")

# Executar o quiz
run_quiz()
```
